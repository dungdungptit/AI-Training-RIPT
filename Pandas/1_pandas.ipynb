## 3.3. Grouping

- Spliting: chia dữ liệu thành các cái nhóm dựa trên một số điều kiện nhất định
- Applying: 1 hàm hoặc nhiều hàm áp dụng cho các nhóm ở trên
- Combining: Kết quả của hàm sau khi áp dụng, kết quả cuối cùng có thể là 1 DataFrame hoặc Series

### 3.3.1. Spliting

```python
df_gr = df.groupby(by='day')
df_gr.head(5)
```

### 3.3.2. Applying

```python
df_gr['total_bill'].mean()
```

```python
df_gr['total_bill'].mean() > 20
```

### 3.3.3. Combining

```python
df_gr_gender_day = df.groupby(by=['gender', 'day'])['total_bill'].sum()
df_gr_gender_day
```

```python
# Aggrigation function
df_gr_gender_day = df.groupby(by=['gender', 'day']).agg({'total_bill': 'sum', 'tip': 'mean'})
df_gr_gender_day
```

```python
# Aggrigation function
df_gr_gender_day = df.groupby(by=['gender', 'day']).agg({'total_bill': 'sum', 'tip': 'mean'}).reset_index()
df_gr_gender_day
```
